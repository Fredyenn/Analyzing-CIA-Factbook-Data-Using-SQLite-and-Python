# Analyzing CIA Factbook Data Using SQLite and Python

The purpose of this project is using SQLite to conduct initial data extraction andd cleaning from a database. Further investigation can be done with other ML models.

The data set is from [CIA World Factbook](https://www.cia.gov/library/publications/the-world-factbook/) the factbookdb can be found in [this repo](https://github.com/factbook/factbook.sql/releases)

Some example of the info this data set contains:

- <font color='red'>population</font>  - The population as of 2015.
- <font color='red'>population_growth</font>  - The annual population growth rate, as a percentage.
- <font color='red'>area</font>  - The total land and water area.


## Load the data and have the first glance of it

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("factbook.db")
cursor = conn.cursor()
cursor.execute("SELECT * FROM sqlite_master").fetchall()
```

```python
q1 = "SELECT * FROM sqlite_master WHERE type='table';"
pd.read_sql_query(q1, conn)
```

```python
# First five instances of the data set
q2 = "select * from facts limit 5"
pd.read_sql_query(q2, conn)
```

```python
# Pull some summary of the table

q3 = '''
select min(population) min_pop, max(population) max_pop, 
min(population_growth) min_pop_grwth, max(population_growth) max_pop_grwth 
from facts
'''
pd.read_sql_query(q3, conn)
```

```python
# One of the instance is the whole world population. Identify it, so it may be useful for further analysis such us normalization.
q4 = '''
select *
from facts
where population == (select max(population) from facts);
'''

pd.read_sql_query(q4, conn)
```

## Histograms

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

q6 = '''
select population, population_growth, birth_rate, death_rate
from facts
where population != (select max(population) from facts)
and population != (select min(population) from facts);
'''
fig , ax= plt.subplots(figsize = (11,11))
pd.read_sql_query(q6, conn).hist(ax=ax)
```

# Population Density of each country

```python
q7 = '''
select name, cast(population as float)/cast(area as float) as density 
from facts 
order by density desc 
limit 15
'''
pd.read_sql_query(q7, conn)
```

## Population density on actual livable area (land)

```python
q7 = '''
select name, cast(population as float)/cast(area_land as float) as density_land_only 
from facts 
order by density_land_only desc 
limit 15
'''
pd.read_sql_query(q7, conn)
```
